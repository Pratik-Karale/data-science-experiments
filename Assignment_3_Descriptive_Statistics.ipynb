## Problem Statement
To compute summary statistics such as mean, median, minimum, maximum, and standard deviation for a dataset grouped by a categorical variable, and to display statistical details including percentiles for each category using Python.

## Theory
**Measures of Central Tendency:**
- Mean: Average value
- Median: Middle value (50th percentile)

**Measures of Variability:**
- Standard Deviation: Spread of data around mean
- Min/Max: Range of values

**Grouped Statistics:** Data grouped by categorical variables to analyze patterns within subgroups.

**Percentiles:** 25% (Q1), 50% (median), 75% (Q3) — help understand data distribution.

```python
# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

print("Libraries imported successfully!")
```

```python
# Step 2: Load the Iris Dataset
iris = load_iris()
print(f"Feature names: {iris.feature_names}")
print(f"Target names: {iris.target_names}")
print(f"Dataset shape: {iris.data.shape}")
```

```python
# Step 3: Convert Dataset into DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print("DataFrame created!")
df.head()
```

```python
# Step 4: Add Species as a Categorical Column
# Map numeric targets (0,1,2) to species names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = pd.Categorical([species_map[t] for t in iris.target])

print("Species column added.")
print(f"Species distribution:{df['species'].value_counts()}")
```

```python
# Step 5: Rename Columns for Clarity
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
print("Columns renamed:")
print(df.columns.tolist())
df.head()
```

```python
# Step 6: Group Data by Species
grouped = df.groupby('species')
print(f"Groups: {list(grouped.groups.keys())}")
print(f"Samples per group:")
print(grouped.size())
```

```python
# Step 7: Compute Summary Statistics for Each Group
print("=" * 80)
for name, group in grouped:
    print(f"{'=' * 40}")
    print(f"  Species: {name.upper()}")
    print(f"{'=' * 40}")
    stats = group.loc[:,df.columns[:-1]].describe().loc[["50%","std","mean"]]
    # stats = group.iloc[:, :4].agg(['mean', 'median', 'min', 'max', 'std'])
    print(stats.round(3))
    print()
```

```python
# Step 8: Get List of Unique Species
species_list = df['species'].unique().tolist()
print("Species list:", species_list)
```

```python
# Step 9 & 10: Detailed Statistics Using describe() for Each Species
for species in species_list:
    subset = df[df['species'] == species].iloc[:, :4]
    print(f"{'=' * 60}")
    print(f"  DETAILED STATISTICS: {species.upper()}")
    print(f"{'=' * 60}")
    print(subset.describe(percentiles=[0.25, 0.5, 0.75]).round(3))
```

```python
# Step 11: Display Overall Summary
print("=" * 60)
print("  OVERALL COMPARISON OF SPECIES (MEAN VALUES)")
print("=" * 60)
overall = grouped.mean()
print(overall.round(3))

print(f"{'=' * 60}")
print("  OVERALL COMPARISON OF SPECIES (MEDIAN VALUES)")
print("=" * 60)
print(grouped.median().round(3))

print(f"{'=' * 60}")
print("  OVERALL COMPARISON OF SPECIES (STD DEVIATION)")
print("=" * 60)
print(grouped.std().round(3))
```

## Dataset Description

- **Records:** 150 samples (50 per species)
- **Features:** Sepal Length, Sepal Width, Petal Length, Petal Width
- **Classes (Target):** Setosa, Versicolor, Virginica

Descriptive statistics computed for the Iris dataset grouped by species. Mean, median, min, max, and standard deviation were calculated for each species group. Detailed percentiles (25%, 50%, 75%) were displayed using describe(). The statistics reveal clear differences across the three Iris species, particularly in petal dimensions.
